```python
import pandas as pd
```

```python
import nltk

from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
nltk.download('subjectivity')
nltk.download('movie_reviews')

sia = SentimentIntensityAnalyzer()
```

```python
data = pd.read_csv("stockerbot-export.csv", on_bad_lines='skip')
data.head()
```

```python
data['symbols'].value_counts()
```

```python
# Example stock: NFLX
nt = data[data['symbols'] == 'NFLX']
ntsorted = nt.sort_values(by='timestamp')
```

```python
compound_scores = []

for index, row in ntsorted.iterrows():
    tweet = row['text']
    analysis = sia.polarity_scores(tweet)
    compound_scores.append(analysis['compound'])

ntsorted['sentiment_score'] = compound_scores
ntsorted
```

```python
import matplotlib.pyplot as plt

plt.plot(ntsorted['timestamp'], compound_scores)
plt.ylabel('Compound Score')

plt.show()
```
